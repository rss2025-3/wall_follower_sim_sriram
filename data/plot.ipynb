```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
wall1 = pd.read_csv("../../../wall_following_metrics.csv")
wall1 = wall1[wall1["timestamp"] < 7.5]
```

```python
plt.figure(figsize=(6,4))
plt.plot(wall1['timestamp'], wall1['side_distance'], label='Right Side Wall Velocity 2')
plt.xlabel('Time (seconds)')
plt.ylabel('Distance from Wall (meters)')
plt.title('Distance From Wall')
plt.grid(True)
plt.legend()
plt.show()
```
